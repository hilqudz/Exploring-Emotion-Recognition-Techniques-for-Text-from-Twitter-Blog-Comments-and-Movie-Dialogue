```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import time
import re
from unicodedata import normalize
import nltk
from collections import Counter

# Download necessary NLTK data
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('punkt_tab')

from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import SnowballStemmer

# --- Sklearn (Machine Learning) ---
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (
    confusion_matrix, classification_report, accuracy_score,
    precision_score, recall_score, f1_score
)

# --- TensorFlow / Keras (Deep Learning - BiLSTM) ---
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Bidirectional, LSTM, Dropout, Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.utils import to_categorical

# --- PyTorch / Transformers (Deep Learning - BERT, etc.) ---
import torch
from torch.optim import AdamW
from torch.utils.data import DataLoader, Dataset
from transformers import (
    BertTokenizer, BertForSequenceClassification,
    RobertaTokenizer, RobertaForSequenceClassification,
    DistilBertTokenizer, DistilBertForSequenceClassification,
    get_linear_schedule_with_warmup
)

# Set random seeds for reproducibility
np.random.seed(42)
tf.random.set_seed(42)
torch.manual_seed(42)

# ====== 1. Memuat Data ======
print("Loading datasets...")
try:
    train_data = pd.read_csv('/kaggle/input/emotion-dataset/training.csv')
    val_data = pd.read_csv('/kaggle/input/emotion-dataset/validation.csv')
    test_data = pd.read_csv('/kaggle/input/emotion-dataset/test.csv')
except FileNotFoundError:
    print("\nError: Pastikan file 'training.csv', 'validation.csv', dan 'test.csv' ada di direktori /content/.")
    print("Silakan upload file dataset ZIP Anda dan unzip di Google Colab.")
    # Exit if files are not found
    exit()

print(f"Original Train data: {train_data.shape}")
print(f"Original Validation data: {val_data.shape}")
print(f"Original Test data: {test_data.shape}")
print("-" * 30)

# ====== 2. Implementasi dan Debug Preprocessing ======

# Definisikan fungsi preprocessing dari user
def preprocessing(text: str) -> str:
    """
    This function takes raw text as input and performs a series of preprocessing steps.
    """
    if not isinstance(text, str):
        return ""
    text = text.lower()
    text = " ".join(w for w in text.split()
                    if 'http://' not in w and 'https://' not in w
                    and '.com' not in w and '.es' not in w and w != 'rt')
    text = re.sub(r'[@#]', '', text)
    text = re.sub(r'\b([jh]*[aeiou]*[jh]+[aeiou]*)*\b', '', text)
    text = re.sub(r'(.)\1{2,}', r'\1', text)
    text = re.sub(r"([^n\u0300-\u036f]|n(?!\u0303(?![\u0300-\u036f])))[\u0300-\u036f]+",
                  r"\1", normalize("NFD", text), flags=re.I)
    text = re.sub("[^a-zA-Z]", " ", text)
    text = re.sub(r"\s+", " ", text).strip()

    tokens = word_tokenize(text)
    stop_words = set(stopwords.words('english'))
    stemmer = SnowballStemmer('english')

    return " ".join(stemmer.stem(w) for w in tokens if w not in stop_words)

# Tampilkan data sebelum preprocessing
print("--- Data SEBELUM Preprocessing ---")
print(train_data.head())
print("-" * 30)

# Simpan data asli untuk perbandingan
train_data_before = train_data.copy()

# Terapkan fungsi preprocessing ke semua dataset
print("Applying preprocessing function...")
train_data['text_cleaned'] = train_data['text'].apply(preprocessing)
val_data['text_cleaned'] = val_data['text'].apply(preprocessing)
test_data['text_cleaned'] = test_data['text'].apply(preprocessing)

# Tampilkan data sesudah preprocessing
print("--- Data SESUDAH Preprocessing ---")

# Atur opsi pandas untuk menampilkan seluruh isi kolom tanpa terpotong
pd.set_option('display.max_colwidth', None)

display_df = pd.DataFrame({
    'Original Text': train_data_before['text'].head(),
    'Cleaned Text': train_data['text_cleaned'].head(),
    'Label': train_data['label'].head()
})

# Gunakan display() untuk output yang lebih rapi di Colab/Jupyter
from IPython.display import display
display(display_df)

# (Opsional) Kembalikan ke pengaturan default jika diperlukan nanti
# pd.reset_option('display.max_colwidth')

print("-" * 30)


# ====== 3. Sebaran Emosi per Dataset ======
labels_dict = {0: 'sadness', 1: 'joy', 2: 'love', 3: 'anger', 4: 'fear', 5: 'surprise'}
train_data['label_name'] = train_data['label'].map(labels_dict)
val_data['label_name'] = val_data['label'].map(labels_dict)
test_data['label_name'] = test_data['label'].map(labels_dict)

fig, axes = plt.subplots(1, 3, figsize=(20, 5))
fig.suptitle('Persebaran Emosi per Dataset', fontsize=16)

sns.countplot(ax=axes[0], x='label_name', data=train_data, order=labels_dict.values())
axes[0].set_title('Training Data')
axes[0].tick_params(axis='x', rotation=45)

sns.countplot(ax=axes[1], x='label_name', data=val_data, order=labels_dict.values())
axes[1].set_title('Validation Data')
axes[1].tick_params(axis='x', rotation=45)

sns.countplot(ax=axes[2], x='label_name', data=test_data, order=labels_dict.values())
axes[2].set_title('Test Data')
axes[2].tick_params(axis='x', rotation=45)

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()


# ====== 4. Model Machine Learning (Naive Bayes, SVM, Random Forest) ======
print("\n" + "="*50)
print("  Training & Evaluating Machine Learning Models")
print("="*50 + "\n")

# Ekstraksi Fitur TF-IDF
X_train = train_data['text_cleaned']
y_train = train_data['label']
X_val = val_data['text_cleaned']
y_val = val_data['label']
X_test = test_data['text_cleaned']
y_test = test_data['label']

print("Extracting features using TF-IDF...")
tfidf_vectorizer = TfidfVectorizer(max_features=5000,ngram_range=(1,2)) # Batasi fitur untuk efisiensi
X_train_tfidf = tfidf_vectorizer.fit_transform(X_train)
X_test_tfidf = tfidf_vectorizer.transform(X_test)
print(f"TF-IDF Matrix Shape: {X_train_tfidf.shape}")

# Definisikan model-model
ml_models = {
    "Naive Bayes": MultinomialNB(),
    "SVM": SVC(kernel='linear'), # Linear kernel lebih cepat untuk data teks
    "Random Forest": RandomForestClassifier(n_estimators=100, random_state=42)
}

# Latih dan evaluasi setiap model
for name, model in ml_models.items():
    print(f"\n--- Training {name} ---")

    # Training
    start_time = time.time()
    model.fit(X_train_tfidf, y_train)

    # Inference
    start_infer_time = time.time()
    y_pred = model.predict(X_test_tfidf)
    inference_time = time.time() - start_infer_time

    # Evaluasi
    accuracy = accuracy_score(y_test, y_pred)
    report = classification_report(y_test, y_pred, target_names=labels_dict.values(), digits=4)
    cm = confusion_matrix(y_test, y_pred)

    print(f"✅ {name} Evaluation:")
    print(f"   Inference Time: {inference_time:.4f} seconds")
    print(f"   Accuracy: {accuracy:.4f}")
    print("\nClassification Report:")
    print(report)

    # Plot Confusion Matrix
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=labels_dict.values(), yticklabels=labels_dict.values())
    plt.title(f'Confusion Matrix - {name}')
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    plt.show()


# ====== 5. Model Deep Learning - BiLSTM ======
print("\n" + "="*50)
print("     Training & Evaluating BiLSTM Model")
print("="*50 + "\n")

# Hyperparameters
VOCAB_SIZE = 10000
EMBEDDING_DIM = 128
MAX_LEN = 128
LEARNING_RATE = 1e-3
EPOCHS_BILSTM = 20

# Tokenisasi dan Padding
tokenizer = Tokenizer(num_words=VOCAB_SIZE, oov_token='<OOV>')
tokenizer.fit_on_texts(train_data['text_cleaned'])

train_sequences = tokenizer.texts_to_sequences(train_data['text_cleaned'])
val_sequences = tokenizer.texts_to_sequences(val_data['text_cleaned'])
test_sequences = tokenizer.texts_to_sequences(test_data['text_cleaned'])

X_train_padded = pad_sequences(train_sequences, maxlen=MAX_LEN, padding='post', truncating='post')
X_val_padded = pad_sequences(val_sequences, maxlen=MAX_LEN, padding='post', truncating='post')
X_test_padded = pad_sequences(test_sequences, maxlen=MAX_LEN, padding='post', truncating='post')

y_train_cat = to_categorical(train_data['label'], num_classes=6)
y_val_cat = to_categorical(val_data['label'], num_classes=6)
y_test_cat = to_categorical(test_data['label'], num_classes=6)

# Pembangunan Model BiLSTM
bilstm_model = Sequential([
    Embedding(input_dim=VOCAB_SIZE, output_dim=EMBEDDING_DIM, input_length=MAX_LEN),
    Bidirectional(LSTM(128, return_sequences=True)),
    Dropout(0.3),
    Bidirectional(LSTM(64)),
    Dropout(0.3),
    Dense(64, activation='relu'),
    Dense(6, activation='softmax')
])

bilstm_model.compile(
    loss='categorical_crossentropy',
    optimizer=Adam(learning_rate=LEARNING_RATE),
    metrics=['accuracy']
)

print(bilstm_model.summary())

# Training
print("\n--- Training BiLSTM ---")
history_bilstm = bilstm_model.fit(
    X_train_padded, y_train_cat,
    epochs=EPOCHS_BILSTM,
    validation_data=(X_val_padded, y_val_cat),
    verbose=1
)

# Plot Akurasi dan Loss
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))
ax1.plot(history_bilstm.history['accuracy'], label='Train Accuracy')
ax1.plot(history_bilstm.history['val_accuracy'], label='Validation Accuracy')
ax1.set_title('BiLSTM: Accuracy vs. Epochs')
ax1.set_xlabel('Epoch')
ax1.set_ylabel('Accuracy')
ax1.legend()

ax2.plot(history_bilstm.history['loss'], label='Train Loss')
ax2.plot(history_bilstm.history['val_loss'], label='Validation Loss')
ax2.set_title('BiLSTM: Loss vs. Epochs')
ax2.set_xlabel('Epoch')
ax2.set_ylabel('Loss')
ax2.legend()
plt.show()

# Evaluasi BiLSTM
print("\n--- Evaluating BiLSTM ---")
start_infer_time = time.time()
y_pred_probs = bilstm_model.predict(X_test_padded)
inference_time = time.time() - start_infer_time

y_pred_bilstm = np.argmax(y_pred_probs, axis=1)
y_true_bilstm = np.argmax(y_test_cat, axis=1)

accuracy_bilstm = accuracy_score(y_true_bilstm, y_pred_bilstm)
report_bilstm = classification_report(y_true_bilstm, y_pred_bilstm, target_names=labels_dict.values(), digits=4)
cm_bilstm = confusion_matrix(y_true_bilstm, y_pred_bilstm)

print("✅ BiLSTM Evaluation:")
print(f"   Inference Time: {inference_time:.4f} seconds")
print(f"   Accuracy: {accuracy_bilstm:.4f}")
print("\nClassification Report:")
print(report_bilstm)

plt.figure(figsize=(8, 6))
sns.heatmap(cm_bilstm, annot=True, fmt='d', cmap='Blues', xticklabels=labels_dict.values(), yticklabels=labels_dict.values())
plt.title('Confusion Matrix - BiLSTM')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```

```python
import torch
from transformers import BertTokenizer, BertForSequenceClassification
from transformers import RobertaTokenizer, RobertaForSequenceClassification
from transformers import DistilBertTokenizer, DistilBertForSequenceClassification
from torch.utils.data import DataLoader, Dataset
from torch.optim import AdamW
from transformers import get_linear_schedule_with_warmup
import time
from sklearn.metrics import accuracy_score, classification_report
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Set random seeds for reproducibility
np.random.seed(42)
torch.manual_seed(42)

# ====== 6. Model Deep Learning - Transformers ======
print("\n" + "="*50)
print("  Training & Evaluating Transformer Models")
print("="*50 + "\n")

# Hyperparameters
MAX_LEN_TRANS = 128
BATCH_SIZE = 16
EPOCHS_TRANS = 10
LEARNING_RATE_TRANS = 2e-5

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")

# Dataset Class
class EmotionDataset(Dataset):
    def __init__(self, texts, labels, tokenizer, max_len):
        self.texts = texts
        self.labels = labels
        self.tokenizer = tokenizer
        self.max_len = max_len

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, item):
        text = str(self.texts[item])
        label = self.labels[item]
        encoding = self.tokenizer.encode_plus(
            text,
            add_special_tokens=True,
            max_length=self.max_len,
            return_token_type_ids=False,
            padding='max_length',
            truncation=True,
            return_attention_mask=True,
            return_tensors='pt',
        )
        return {
            'text': text,
            'input_ids': encoding['input_ids'].flatten(),
            'attention_mask': encoding['attention_mask'].flatten(),
            'labels': torch.tensor(label, dtype=torch.long)
        }

def create_data_loader(df, tokenizer, max_len, batch_size):
    ds = EmotionDataset(
        texts=df.text_cleaned.to_numpy(),
        labels=df.label.to_numpy(),
        tokenizer=tokenizer,
        max_len=max_len
    )
    return DataLoader(ds, batch_size=batch_size, num_workers=2)

# Training & Evaluation Loop
def train_epoch(model, data_loader, optimizer, device, scheduler):
    model = model.train()
    losses, correct_predictions = [], 0
    for d in data_loader:
        input_ids = d["input_ids"].to(device)
        attention_mask = d["attention_mask"].to(device)
        labels = d["labels"].to(device)

        outputs = model(input_ids=input_ids, attention_mask=attention_mask, labels=labels)
        loss = outputs.loss
        logits = outputs.logits

        _, preds = torch.max(logits, dim=1)
        correct_predictions += torch.sum(preds == labels)
        losses.append(loss.item())

        loss.backward()
        torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
        optimizer.step()
        scheduler.step()
        optimizer.zero_grad()

    return correct_predictions.double() / len(data_loader.dataset), np.mean(losses)

def eval_model(model, data_loader, device):
    model = model.eval()
    losses, correct_predictions = [], 0
    with torch.no_grad():
        for d in data_loader:
            input_ids = d["input_ids"].to(device)
            attention_mask = d["attention_mask"].to(device)
            labels = d["labels"].to(device)

            outputs = model(input_ids=input_ids, attention_mask=attention_mask, labels=labels)
            loss = outputs.loss
            logits = outputs.logits

            _, preds = torch.max(logits, dim=1)
            correct_predictions += torch.sum(preds == labels)
            losses.append(loss.item())

    return correct_predictions.double() / len(data_loader.dataset), np.mean(losses)

# Transformer Models to Run
transformer_models = {
    "BERT": ("bert-base-uncased", BertForSequenceClassification, BertTokenizer),
    "RoBERTa": ("roberta-base", RobertaForSequenceClassification, RobertaTokenizer),
    "DistilBERT": ("distilbert-base-uncased", DistilBertForSequenceClassification, DistilBertTokenizer)
}

# Assuming train_data, val_data, and test_data DataFrames are already loaded and preprocessed
# (as they were in the previous successful cell execution)

# Labels dictionary (assuming it's defined elsewhere, e.g., in the previous cell)
labels_dict = {0: 'sadness', 1: 'joy', 2: 'love', 3: 'anger', 4: 'fear', 5: 'surprise'}


for model_name, (model_hub_name, ModelClass, TokenizerClass) in transformer_models.items():
    print(f"\n--- Running {model_name} ---")

    # Tokenizer and DataLoaders
    tokenizer = TokenizerClass.from_pretrained(model_hub_name)
    train_loader = create_data_loader(train_data, tokenizer, MAX_LEN_TRANS, BATCH_SIZE)
    val_loader = create_data_loader(val_data, tokenizer, MAX_LEN_TRANS, BATCH_SIZE)
    test_loader = create_data_loader(test_data, tokenizer, MAX_LEN_TRANS, BATCH_SIZE)


    # Model Setup
    model = ModelClass.from_pretrained(model_hub_name, num_labels=6)
    model = model.to(device)


    optimizer = AdamW(model.parameters(), lr=LEARNING_RATE_TRANS)
    total_steps = len(train_loader) * EPOCHS_TRANS
    scheduler = get_linear_schedule_with_warmup(
        optimizer,
        num_warmup_steps=0,
        num_training_steps=total_steps
    )

    history = {'train_acc': [], 'train_loss': [], 'val_acc': [], 'val_loss': []}

    # Training Loop
    for epoch in range(EPOCHS_TRANS):
        print(f'Epoch {epoch + 1}/{EPOCHS_TRANS}')
        train_acc, train_loss = train_epoch(model, train_loader, optimizer, device, scheduler)
        print(f'Train loss {train_loss:.4f}, accuracy {train_acc:.4f}')

        val_acc, val_loss = eval_model(model, val_loader, device)
        print(f'Val loss {val_loss:.4f}, accuracy {val_acc:.4f}\n')

        history['train_acc'].append(train_acc)
        history['train_loss'].append(train_loss)
        history['val_acc'].append(val_acc)
        history['val_loss'].append(val_loss)

    # Plot Akurasi dan Loss
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))
    ax1.plot([h.cpu().numpy() for h in history['train_acc']], label='Train Accuracy')
    ax1.plot([h.cpu().numpy() for h in history['val_acc']], label='Validation Accuracy')
    ax1.set_title(f'{model_name}: Accuracy vs. Epochs')
    ax1.legend()

    ax2.plot([h for h in history['train_loss']], label='Train Loss') # Loss is already a float
    ax2.plot([h for h in history['val_loss']], label='Validation Loss') # Loss is already a float
    ax2.set_title(f'{model_name}: Loss vs. Epochs')
    ax2.legend()
    plt.show()

    # Evaluation
    print(f"\n--- Evaluating {model_name} on Test Set ---")

    y_review_texts, y_pred, y_test_labels = [], [], []

    start_infer_time = time.time()
    model.eval()
    with torch.no_grad():
        for d in test_loader:
            texts = d["text"]
            input_ids = d["input_ids"].to(device)
            attention_mask = d["attention_mask"].to(device)
            labels = d["labels"].to(device)

            outputs = model(input_ids=input_ids, attention_mask=attention_mask, labels=labels)
            _, preds = torch.max(outputs.logits, dim=1)

            y_review_texts.extend(texts)
            y_pred.extend(preds)
            y_test_labels.extend(labels)

    inference_time = time.time() - start_infer_time

    y_pred = torch.stack(y_pred).cpu().numpy()
    y_test_labels = torch.stack(y_test_labels).cpu().numpy()

    print("✅ {} Evaluation:".format(model_name))
    print(f"   Inference Time: {inference_time:.4f} seconds")
    print(f"   Accuracy: {accuracy_score(y_test_labels, y_pred):.4f}")
    print("\nClassification Report:")
    print(classification_report(y_test_labels, y_pred, target_names=labels_dict.values(), digits=4))

    # Confusion Matrix
    cm = confusion_matrix(y_test_labels, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=labels_dict.values(), yticklabels=labels_dict.values())
    plt.title(f'Confusion Matrix - {model_name}')
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    plt.show()

print("\n" + "="*50)
print("             All Models Evaluated!")
print("="*50 + "\n")
```
