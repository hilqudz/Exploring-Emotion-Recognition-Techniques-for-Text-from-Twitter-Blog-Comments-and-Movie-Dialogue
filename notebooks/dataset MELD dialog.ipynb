```python
import torch
import torch.nn as nn
from transformers import BertTokenizer, BertForSequenceClassification, get_linear_schedule_with_warmup, \
                     RobertaTokenizer, RobertaForSequenceClassification, DistilBertTokenizer, DistilBertForSequenceClassification
from torch.utils.data import Dataset, DataLoader
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
import seaborn as sns
import time
from torch.optim import AdamW
from IPython.display import display # BARU: Untuk menampilkan dataframe dengan baik

# For Traditional ML models
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
import random # For set_seed

# Set device
device = torch.device('cuda') if torch.cuda.is_available() else torch.device('cpu')
print(f"Using device: {device}")

# Set seed for reproducibility
def set_seed(seed_value=42):
    random.seed(seed_value)
    np.random.seed(seed_value)
    torch.manual_seed(seed_value)
    if torch.cuda.is_available():
        torch.cuda.manual_seed_all(seed_value)
        torch.backends.cudnn.deterministic = True
        torch.backends.cudnn.benchmark = False

set_seed()

# BARU: Fungsi untuk menyimpan hasil prediksi ke file CSV
def save_prediction_results(test_dataframe, y_true, y_pred, id_map, model_name):
    """Menyimpan hasil prediksi ke file CSV untuk analisis kesalahan."""
    # Buat salinan untuk menghindari modifikasi dataframe asli
    results_df = test_dataframe.copy().reset_index(drop=True)
    
    # Konversi Series ke array jika perlu
    if isinstance(y_true, pd.Series):
        y_true = y_true.values

    # Tambahkan label asli dan prediksi (yang dapat dibaca manusia)
    results_df['true_emotion'] = [id_map.get(label, 'N/A') for label in y_true]
    results_df['predicted_emotion'] = [id_map.get(label, 'N/A') for label in y_pred]
    results_df['is_correct'] = (results_df['true_emotion'] == results_df['predicted_emotion'])
    
    # Pilih kolom yang relevan untuk output
    output_df = results_df[['Utterance', 'true_emotion', 'predicted_emotion', 'is_correct']]
    
    # Hasilkan nama file dan simpan
    filename = f"prediction_results_{model_name.replace(' ', '_').replace('(', '').replace(')', '')}.csv"
    output_df.to_csv(filename, index=False)
    
    print(f"\n✅ Prediction results for {model_name} saved to '{filename}'")
    
    # Tampilkan sampel entri yang salah diklasifikasikan
    misclassified = output_df[~output_df['is_correct']]
    if not misclassified.empty:
        print(f"--- Contoh Kesalahan Prediksi ({model_name}) ---")
        display(misclassified.head())


# Load dataset
try:
    train_df = pd.read_csv('/kaggle/input/melddataset/train_sent_emo.csv')
    val_df = pd.read_csv('/kaggle/input/melddataset/dev_sent_emo.csv')
    test_df = pd.read_csv('/kaggle/input/melddataset/test_sent_emo.csv')
    print("MELD dataset loaded successfully.")
except FileNotFoundError:
    print("MELD Dataset files not found. Using dummy data for demonstration.")
    data = {'Utterance': ['I am happy', 'This is sad', 'Feeling angry now', 'Quite surprised by this', 'It is neutral news', 'Feeling joyful today', 'He seems afraid', 'She is disgusted by it'] * 1000,
            'Emotion': ['joy', 'sadness', 'anger', 'surprise', 'neutral', 'joy', 'fear', 'disgust'] * 1000}
    full_df = pd.DataFrame(data)
    train_df = full_df.sample(frac=0.8, random_state=42).reset_index(drop=True)
    val_df = full_df.sample(frac=0.1, random_state=1).reset_index(drop=True)
    test_df = full_df.sample(frac=0.1, random_state=2).reset_index(drop=True)

# Select relevant columns and ensure 'Utterance' is string
train_df = train_df[['Utterance', 'Emotion']].copy()
train_df['Utterance'] = train_df['Utterance'].astype(str)
val_df = val_df[['Utterance', 'Emotion']].copy()
val_df['Utterance'] = val_df['Utterance'].astype(str)
test_df = test_df[['Utterance', 'Emotion']].copy()
test_df['Utterance'] = test_df['Utterance'].astype(str)

# Label encoding
all_emotions = pd.concat([train_df['Emotion'], val_df['Emotion'], test_df['Emotion']]).astype(str).unique()
emotion_labels = {label: idx for idx, label in enumerate(all_emotions)}
id_to_emotion_map = {v: k for k, v in emotion_labels.items()} # BARU: Membuat map dari id ke label
print("\nEmotion labels mapping:", emotion_labels)
num_labels = len(emotion_labels)
target_names_list = [k for k, v in sorted(emotion_labels.items(), key=lambda item: item[1])]
print(f"\nNumber of unique emotion classes: {num_labels}")

train_df['Emotion_Encoded'] = train_df['Emotion'].astype(str).map(emotion_labels).astype(int)
val_df['Emotion_Encoded'] = val_df['Emotion'].astype(str).map(emotion_labels).astype(int)
test_df['Emotion_Encoded'] = test_df['Emotion'].astype(str).map(emotion_labels).astype(int)

# (Sisa kode dari Dataset Class, DataLoader, dan fungsi training/evaluasi tetap sama)
# ...
# ... Bagian ini tidak diubah ...
# ...

# --- Fungsi Helper Baru untuk Confusion Matrix ---
def plot_confusion_matrix(y_true, y_pred, labels, model_name):
    """Plots a confusion matrix using seaborn."""
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=labels, yticklabels=labels)
    plt.title(f'Confusion Matrix - {model_name}', fontsize=16)
    plt.xlabel('Predicted Label', fontsize=12)
    plt.ylabel('True Label', fontsize=12)
    plt.show()

# Dataset Class (common for PyTorch models)
class SentimentDataset(Dataset):
    def __init__(self, texts, labels, tokenizer, max_len):
        self.texts = texts
        self.labels = labels
        self.tokenizer = tokenizer
        self.max_len = max_len

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, index):
        text = str(self.texts[index])
        label = self.labels[index]
        encoding = self.tokenizer.encode_plus(
            text,
            add_special_tokens=True,
            max_length=self.max_len,
            padding='max_length',
            truncation=True,
            return_attention_mask=True,
            return_tensors='pt'
        )
        return {
            'text': text,
            'input_ids': encoding['input_ids'].flatten(),
            'attention_mask': encoding['attention_mask'].flatten(),
            'label': torch.tensor(label, dtype=torch.long)
        }

# Function to create DataLoader (common for PyTorch models)
def create_pytorch_data_loader(df, text_column, label_column, tokenizer, max_len, batch_size, shuffle=False):
    dataset = SentimentDataset(
        texts=df[text_column].to_numpy(),
        labels=df[label_column].to_numpy(),
        tokenizer=tokenizer,
        max_len=max_len
    )
    return DataLoader(dataset, batch_size=batch_size, num_workers=0, pin_memory=True, shuffle=shuffle)

# --- Transformer Model Training/Evaluation Functions ---
def train_epoch_hf_loss(model, data_loader, optimizer, scheduler, device):
    model.train()
    total_loss, correct_preds, total_samples = 0, 0, 0
    for data in data_loader:
        input_ids = data['input_ids'].to(device)
        attention_mask = data['attention_mask'].to(device)
        labels = data['label'].to(device)

        optimizer.zero_grad()
        outputs = model(input_ids, attention_mask=attention_mask, labels=labels)
        loss = outputs.loss
        logits = outputs.logits
        loss.backward()
        optimizer.step()
        if scheduler:
            scheduler.step()

        total_loss += loss.item()
        _, preds = torch.max(logits, 1)
        correct_preds += (preds == labels).sum().item()
        total_samples += labels.size(0)
    return total_loss / len(data_loader), correct_preds / total_samples

def eval_model_hf_loss(model, data_loader, device):
    model.eval()
    total_loss, correct_preds, total_samples = 0, 0, 0
    all_preds_list = []
    all_labels_list = []
    with torch.no_grad():
        for data in data_loader:
            input_ids = data['input_ids'].to(device)
            attention_mask = data['attention_mask'].to(device)
            labels = data['label'].to(device)

            outputs = model(input_ids, attention_mask=attention_mask, labels=labels)
            loss = outputs.loss
            logits = outputs.logits
            total_loss += loss.item()
            _, preds = torch.max(logits, 1)

            all_preds_list.extend(preds.cpu().numpy())
            all_labels_list.extend(labels.cpu().numpy())
            correct_preds += (preds == labels).sum().item()
            total_samples += labels.size(0)
    return total_loss / len(data_loader), correct_preds / total_samples, all_labels_list, all_preds_list

def inference_time_pytorch(model, data_loader, device):
    model.eval()
    start_time = time.time()
    with torch.no_grad():
        for batch in data_loader:
            input_ids = batch["input_ids"].to(device)
            attention_mask = batch["attention_mask"].to(device)
            model(input_ids=input_ids, attention_mask=attention_mask)
    total_time = time.time() - start_time
    return total_time, total_time / len(data_loader.dataset)

def train_and_evaluate_transformer(model, train_data_loader, val_data_loader, test_data_loader,
                                   optimizer, scheduler, device, model_name, epochs, emotion_labels_map):
    print(f"\n--- Training and Evaluating {model_name} ---")
    best_val_acc = 0
    best_model_state_dict = None

    for epoch_num in range(epochs):
        print(f'{model_name} Epoch {epoch_num + 1}/{epochs}')
        train_loss, train_acc = train_epoch_hf_loss(model, train_data_loader, optimizer, scheduler, device)
        print(f'Train loss: {train_loss:.4f}, Train accuracy: {train_acc:.4f}')

        val_loss, val_acc, _, _ = eval_model_hf_loss(model, val_data_loader, device)
        print(f'Validation loss: {val_loss:.4f}, Validation accuracy: {val_acc:.4f}')

        if val_acc > best_val_acc:
            best_val_acc = val_acc
            best_model_state_dict = model.state_dict().copy()
        print("-" * 30)

    if best_model_state_dict:
        model.load_state_dict(best_model_state_dict)

    test_loss, test_acc, true_labels, predictions = eval_model_hf_loss(model, test_data_loader, device)
    print(f'{model_name} Test loss: {test_loss:.4f}, Test accuracy: {test_acc:.4f}')

    print(f'\n{model_name} Classification Report on Test Set:')
    print(classification_report(true_labels, predictions, target_names=target_names_list, zero_division=0))

    total_time, avg_time_per_sample = inference_time_pytorch(model, test_data_loader, device)
    print(f'{model_name} Inference time on Test Set: {total_time:.3f}s total / {avg_time_per_sample:.6f}s per sample')
    
    # BARU: Menyimpan hasil prediksi Transformer
    save_prediction_results(test_df, true_labels, predictions, id_to_emotion_map, model_name)

    print(f"\nDisplaying Confusion Matrix for {model_name}...")
    plot_confusion_matrix(true_labels, predictions, target_names_list, model_name)


# --- Transformer Models Setup & Training ---
# Hyperparameters for Transformers
MAX_LEN_TRANSFORMER = 128
BATCH_SIZE_TRANSFORMER = 16
EPOCHS_TRANSFORMER = 10
LEARNING_RATE_TRANSFORMER = 2e-5

# BERT
bert_tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
train_dl_bert = create_pytorch_data_loader(train_df, 'Utterance', 'Emotion_Encoded', bert_tokenizer, MAX_LEN_TRANSFORMER, BATCH_SIZE_TRANSFORMER, shuffle=True)
val_dl_bert = create_pytorch_data_loader(val_df, 'Utterance', 'Emotion_Encoded', bert_tokenizer, MAX_LEN_TRANSFORMER, BATCH_SIZE_TRANSFORMER)
test_dl_bert = create_pytorch_data_loader(test_df, 'Utterance', 'Emotion_Encoded', bert_tokenizer, MAX_LEN_TRANSFORMER, BATCH_SIZE_TRANSFORMER)
model_bert = BertForSequenceClassification.from_pretrained('bert-base-uncased', num_labels=num_labels).to(device)
optimizer_bert = AdamW(model_bert.parameters(), lr=LEARNING_RATE_TRANSFORMER)
scheduler_bert = get_linear_schedule_with_warmup(optimizer_bert, num_warmup_steps=0, num_training_steps=len(train_dl_bert) * EPOCHS_TRANSFORMER)
train_and_evaluate_transformer(model_bert, train_dl_bert, val_dl_bert, test_dl_bert, optimizer_bert, scheduler_bert, device, 'BERT', EPOCHS_TRANSFORMER, emotion_labels)

# RoBERTa
roberta_tokenizer = RobertaTokenizer.from_pretrained('roberta-base')
train_dl_roberta = create_pytorch_data_loader(train_df, 'Utterance', 'Emotion_Encoded', roberta_tokenizer, MAX_LEN_TRANSFORMER, BATCH_SIZE_TRANSFORMER, shuffle=True)
val_dl_roberta = create_pytorch_data_loader(val_df, 'Utterance', 'Emotion_Encoded', roberta_tokenizer, MAX_LEN_TRANSFORMER, BATCH_SIZE_TRANSFORMER)
test_dl_roberta = create_pytorch_data_loader(test_df, 'Utterance', 'Emotion_Encoded', roberta_tokenizer, MAX_LEN_TRANSFORMER, BATCH_SIZE_TRANSFORMER)
model_roberta = RobertaForSequenceClassification.from_pretrained('roberta-base', num_labels=num_labels).to(device)
optimizer_roberta = AdamW(model_roberta.parameters(), lr=LEARNING_RATE_TRANSFORMER)
scheduler_roberta = get_linear_schedule_with_warmup(optimizer_roberta, num_warmup_steps=0, num_training_steps=len(train_dl_roberta) * EPOCHS_TRANSFORMER)
train_and_evaluate_transformer(model_roberta, train_dl_roberta, val_dl_roberta, test_dl_roberta, optimizer_roberta, scheduler_roberta, device, 'RoBERTa', EPOCHS_TRANSFORMER, emotion_labels)

# DistilBERT
distilbert_tokenizer = DistilBertTokenizer.from_pretrained('distilbert-base-uncased')
train_dl_distilbert = create_pytorch_data_loader(train_df, 'Utterance', 'Emotion_Encoded', distilbert_tokenizer, MAX_LEN_TRANSFORMER, BATCH_SIZE_TRANSFORMER, shuffle=True)
val_dl_distilbert = create_pytorch_data_loader(val_df, 'Utterance', 'Emotion_Encoded', distilbert_tokenizer, MAX_LEN_TRANSFORMER, BATCH_SIZE_TRANSFORMER)
test_dl_distilbert = create_pytorch_data_loader(test_df, 'Utterance', 'Emotion_Encoded', distilbert_tokenizer, MAX_LEN_TRANSFORMER, BATCH_SIZE_TRANSFORMER)
model_distilbert = DistilBertForSequenceClassification.from_pretrained('distilbert-base-uncased', num_labels=num_labels).to(device)
optimizer_distilbert = AdamW(model_distilbert.parameters(), lr=LEARNING_RATE_TRANSFORMER)
scheduler_distilbert = get_linear_schedule_with_warmup(optimizer_distilbert, num_warmup_steps=0, num_training_steps=len(train_dl_distilbert) * EPOCHS_TRANSFORMER)
train_and_evaluate_transformer(model_distilbert, train_dl_distilbert, val_dl_distilbert, test_dl_distilbert, optimizer_distilbert, scheduler_distilbert, device, 'DistilBERT', EPOCHS_TRANSFORMER, emotion_labels)


# --- BiLSTM Model (Keras) ---
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Bidirectional, LSTM, Dropout, Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.callbacks import ModelCheckpoint

# Hyperparameters BiLSTM
VOCAB_SIZE_BILSTM = 10000
EMBED_DIM_BILSTM = 128
MAX_LEN_BILSTM = 128
LEARNING_RATE_BILSTM = 1e-3
EPOCHS_BILSTM = 20
BATCH_SIZE_BILSTM = 64

# 1. Tokenisasi dan Padding Data untuk Keras
print("\n--- BiLSTM (Keras) Setup ---")
print("Tokenizing data for BiLSTM...")
keras_tokenizer = Tokenizer(num_words=VOCAB_SIZE_BILSTM, oov_token="<unk>")
keras_tokenizer.fit_on_texts(train_df['Utterance'])
train_sequences = keras_tokenizer.texts_to_sequences(train_df['Utterance'])
val_sequences = keras_tokenizer.texts_to_sequences(val_df['Utterance'])
test_sequences = keras_tokenizer.texts_to_sequences(test_df['Utterance'])
X_train_pad = pad_sequences(train_sequences, maxlen=MAX_LEN_BILSTM, padding='post', truncating='post')
X_val_pad = pad_sequences(val_sequences, maxlen=MAX_LEN_BILSTM, padding='post', truncating='post')
X_test_pad = pad_sequences(test_sequences, maxlen=MAX_LEN_BILSTM, padding='post', truncating='post')
y_train, y_val, y_test = train_df['Emotion_Encoded'].values, val_df['Emotion_Encoded'].values, test_df['Emotion_Encoded'].values

# 2. Pembangunan Model BiLSTM
model_bilstm = Sequential([
    Embedding(input_dim=VOCAB_SIZE_BILSTM, output_dim=EMBED_DIM_BILSTM, input_length=MAX_LEN_BILSTM),
    Bidirectional(LSTM(128, return_sequences=True)),
    Dropout(0.3),
    Bidirectional(LSTM(64)),
    Dropout(0.3),
    Dense(64, activation='relu'),
    Dense(num_labels, activation='softmax')
])
model_bilstm.compile(optimizer=Adam(learning_rate=LEARNING_RATE_BILSTM), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model_bilstm.summary()

# 4. Training Model
print(f"\nTraining BiLSTM for {EPOCHS_BILSTM} epochs...")
checkpoint = ModelCheckpoint("bilstm_best_model.h5", monitor='val_accuracy', save_best_only=True, mode='max', verbose=1)
history_bilstm = model_bilstm.fit(X_train_pad, y_train, epochs=EPOCHS_BILSTM, batch_size=BATCH_SIZE_BILSTM, validation_data=(X_val_pad, y_val), callbacks=[checkpoint], verbose=1)

# 5. Evaluasi Model
print("\n--- Evaluating BiLSTM Model ---")
model_bilstm.load_weights("bilstm_best_model.h5")
start_time = time.time()
pred_probs = model_bilstm.predict(X_test_pad)
inference_time_bilstm = time.time() - start_time
y_pred_bilstm = np.argmax(pred_probs, axis=1)

print("\nBiLSTM Classification Report on Test Set:")
print(classification_report(y_test, y_pred_bilstm, target_names=target_names_list, digits=4))
print(f"BiLSTM Inference time: {inference_time_bilstm:.4f}s total / {inference_time_bilstm/len(y_test):.6f}s per sample")

# BARU: Menyimpan hasil prediksi BiLSTM
save_prediction_results(test_df, y_test, y_pred_bilstm, id_to_emotion_map, "BiLSTM_Keras")

print("\nDisplaying Confusion Matrix for BiLSTM...")
plot_confusion_matrix(y_test, y_pred_bilstm, target_names_list, "BiLSTM (Keras)")


# --- Traditional ML Models (NB, SVM, RF) ---
print("\n--- Setting up Traditional ML Models ---")
X_train_text = train_df['Utterance']
y_train_num = train_df['Emotion_Encoded']
X_test_text = test_df['Utterance']
y_test_num = test_df['Emotion_Encoded']

print("Vectorizing text data with TF-IDF...")
tfidf_vectorizer = TfidfVectorizer(max_features=5000, ngram_range=(1,2), stop_words='english')
X_train_tfidf = tfidf_vectorizer.fit_transform(X_train_text)
X_test_tfidf = tfidf_vectorizer.transform(X_test_text)

# Naive Bayes
print("\n--- Naive Bayes ---")
nb_model = MultinomialNB()
nb_model.fit(X_train_tfidf, y_train_num)
start_time = time.time()
nb_preds = nb_model.predict(X_test_tfidf)
nb_total_time = time.time() - start_time
print("Naive Bayes Classification Report:")
print(classification_report(y_test_num, nb_preds, target_names=target_names_list, zero_division=0))
print(f"NB Accuracy: {accuracy_score(y_test_num, nb_preds):.4f}")
print(f"NB Inference time: {nb_total_time:.3f}s total / {nb_total_time/len(y_test_num):.6f}s per sample")
# BARU: Menyimpan hasil prediksi Naive Bayes
save_prediction_results(test_df, y_test_num, nb_preds, id_to_emotion_map, "Naive_Bayes")
print("\nDisplaying Confusion Matrix for Naive Bayes...")
plot_confusion_matrix(y_test_num, nb_preds, target_names_list, "Naive Bayes")

# SVM
print("\n--- SVM (Linear Kernel) ---")
svm_model = SVC(kernel='linear', C=1.0, random_state=42)
svm_model.fit(X_train_tfidf, y_train_num)
start_time = time.time()
svm_preds = svm_model.predict(X_test_tfidf)
svm_total_time = time.time() - start_time
print("SVM Classification Report:")
print(classification_report(y_test_num, svm_preds, target_names=target_names_list, zero_division=0))
print(f"SVM Accuracy: {accuracy_score(y_test_num, svm_preds):.4f}")
print(f"SVM Inference time: {svm_total_time:.3f}s total / {svm_total_time/len(y_test_num):.6f}s per sample")
# BARU: Menyimpan hasil prediksi SVM
save_prediction_results(test_df, y_test_num, svm_preds, id_to_emotion_map, "SVM")
print("\nDisplaying Confusion Matrix for SVM...")
plot_confusion_matrix(y_test_num, svm_preds, target_names_list, "SVM")


# Random Forest
print("\n--- Random Forest ---")
rf_model = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1)
rf_model.fit(X_train_tfidf, y_train_num)
start_time = time.time()
rf_preds = rf_model.predict(X_test_tfidf)
rf_total_time = time.time() - start_time
print("Random Forest Classification Report:")
print(classification_report(y_test_num, rf_preds, target_names=target_names_list, zero_division=0))
print(f"RF Accuracy: {accuracy_score(y_test_num, rf_preds):.4f}")
print(f"Random Forest Inference time: {rf_total_time:.3f}s total / {rf_total_time/len(y_test_num):.6f}s per sample")
# BARU: Menyimpan hasil prediksi Random Forest
save_prediction_results(test_df, y_test_num, rf_preds, id_to_emotion_map, "Random_Forest")
print("\nDisplaying Confusion Matrix for Random Forest...")
plot_confusion_matrix(y_test_num, rf_preds, target_names_list, "Random Forest")

print("\n--- All models processed. ---")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import time
import re
from unicodedata import normalize
import nltk
from collections import Counter
import random
from IPython.display import display

# ===== 0. Setup Awal dan Fungsi Preprocessing =====

# Download necessary NLTK data
print("Downloading NLTK data...")
nltk.download('stopwords', quiet=True)
nltk.download('punkt', quiet=True)
nltk.download('punkt_tab', quiet=True)

from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import SnowballStemmer

# --- Impor Library Machine Learning & Deep Learning ---
# Sklearn
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score

# TensorFlow/Keras (Hanya untuk Tokenizer BiLSTM)
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

# PyTorch/Transformers
import torch
import torch.nn as nn
from torch.utils.data import DataLoader, Dataset
from torch.optim import AdamW
from transformers import (
    BertTokenizer, BertForSequenceClassification,
    RobertaTokenizer, RobertaForSequenceClassification,
    DistilBertTokenizer, DistilBertForSequenceClassification,
    get_linear_schedule_with_warmup
)

# Set seed untuk reproduktifitas
def set_seed(seed_value=42):
    random.seed(seed_value)
    np.random.seed(seed_value)
    torch.manual_seed(seed_value)
    tf.random.set_seed(seed_value)
    if torch.cuda.is_available():
        torch.cuda.manual_seed_all(seed_value)
        torch.backends.cudnn.deterministic = True
        torch.backends.cudnn.benchmark = False
set_seed()

# Set device
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"✅ Using device: {device}")

# Definisikan fungsi preprocessing dari user
def preprocessing(text: str) -> str:
    """
    Fungsi ini melakukan serangkaian langkah preprocessing pada teks.
    """
    if not isinstance(text, str):
        return ""
    text = text.lower()
    text = " ".join(w for w in text.split()
                    if 'http://' not in w and 'https://' not in w
                    and '.com' not in w and '.es' not in w and w != 'rt')
    text = re.sub(r'[@#]', '', text)
    text = re.sub(r'\b([jh]*[aeiou]*[jh]+[aeiou]*)*\b', '', text)
    text = re.sub(r'(.)\1{2,}', r'\1', text)
    text = re.sub(r"([^n\u0300-\u036f]|n(?!\u0303(?![\u0300-\u036f])))[\u0300-\u036f]+",
                  r"\1", normalize("NFD", text), flags=re.I)
    text = re.sub("[^a-zA-Z]", " ", text)
    text = re.sub(r"\s+", " ", text).strip()
    
    tokens = word_tokenize(text)
    stop_words = set(stopwords.words('english'))
    stemmer = SnowballStemmer('english')
    
    return " ".join(stemmer.stem(w) for w in tokens if w not in stop_words)

def save_prediction_results(test_dataframe, y_true, y_pred, id_map, model_name):
    """
    Menyimpan hasil prediksi ke file CSV untuk analisis kesalahan.
    """
    # Membuat salinan untuk menghindari modifikasi DataFrame asli
    results_df = test_dataframe.copy().reset_index(drop=True)
    
    # Menambahkan label asli dan prediksi (yang dapat dibaca manusia)
    # Memastikan y_true adalah list atau array, bukan Series dengan indeks yang berbeda
    if isinstance(y_true, pd.Series):
        y_true = y_true.values
        
    results_df['true_emotion'] = [id_map[label] for label in y_true]
    results_df['predicted_emotion'] = [id_map[label] for label in y_pred]
    
    # Memeriksa apakah prediksi benar
    results_df['is_correct'] = (results_df['true_emotion'] == results_df['predicted_emotion'])
    
    # Memilih dan menyusun ulang kolom untuk kejelasan
    output_df = results_df[['text', 'text_cleaned', 'true_emotion', 'predicted_emotion', 'is_correct']]
    
    # Menghasilkan nama file dan menyimpan
    filename = f"prediction_results_{model_name.replace(' ', '_').replace('/', '_')}.csv"
    output_df.to_csv(filename, index=False)
    
    print(f"\n✅ Prediction results for {model_name} saved to '{filename}'")
    
    # Menampilkan sampel entri yang salah diklasifikasikan
    misclassified = output_df[output_df['is_correct'] == False]
    if not misclassified.empty:
        print(f"--- Contoh Kesalahan Prediksi ({model_name}) ---")
        display(misclassified.head())
        
# ===== 1. Memuat dan Memproses Data =====
print("\nLoading MELD dataset...")
try:
    train_df = pd.read_csv('/kaggle/input/melddataset/train_sent_emo.csv')
    val_df = pd.read_csv('/kaggle/input/melddataset/dev_sent_emo.csv')
    test_df = pd.read_csv('/kaggle/input/melddataset/test_sent_emo.csv')
    print("✅ MELD dataset loaded successfully.")
except FileNotFoundError:
    print("❌ Error: Pastikan file 'train_sent_emo.csv', 'dev_sent_emo.csv', dan 'test_sent_emo.csv' ada di direktori.")
    exit()

# Membersihkan kolom yang tidak perlu dan memastikan tipe data string
for df in [train_df, val_df, test_df]:
    df.rename(columns={'Utterance': 'text', 'Emotion': 'emotion'}, inplace=True)
    df['text'] = df['text'].astype(str)
    df['emotion'] = df['emotion'].astype(str)

# --- Debug Preprocessing ---
print("\n--- Debugging Preprocessing Step ---")
# Simpan data asli untuk perbandingan
train_df_before = train_df.copy()

# Terapkan fungsi preprocessing
print("Applying preprocessing function to all datasets...")
train_df['text_cleaned'] = train_df['text'].apply(preprocessing)
val_df['text_cleaned'] = val_df['text'].apply(preprocessing)
test_df['text_cleaned'] = test_df['text'].apply(preprocessing)
print("✅ Preprocessing complete.")

# Tampilkan perbandingan sebelum dan sesudah (full text)
pd.set_option('display.max_colwidth', None)
comparison_df = pd.DataFrame({
    'Original Text': train_df_before.head(5)['text'],
    'Cleaned Text': train_df.head(5)['text_cleaned'],
    'Emotion': train_df.head(5)['emotion']
})
print("\nComparison of first 5 rows (Before vs. After Preprocessing):")
display(comparison_df)
print("-" * 50)

# --- Label Encoding ---
all_emotions = pd.concat([train_df['emotion'], val_df['emotion'], test_df['emotion']]).unique()
emotion_to_id = {label: i for i, label in enumerate(all_emotions)}
id_to_emotion = {i: label for label, i in emotion_to_id.items()}
num_labels = len(emotion_to_id)

train_df['label'] = train_df['emotion'].map(emotion_to_id)
val_df['label'] = val_df['emotion'].map(emotion_to_id)
test_df['label'] = test_df['emotion'].map(emotion_to_id)

print(f"\n✅ Emotion mapping created with {num_labels} classes: {emotion_to_id}")
print("-" * 50)


# ===== 2. Model Machine Learning (Naive Bayes, SVM, Random Forest) =====
print("\n===== 2. Training Machine Learning Models =====")

# Ekstraksi Fitur TF-IDF dari teks yang sudah dibersihkan
X_train_ml = train_df['text_cleaned']
y_train_ml = train_df['label']
X_test_ml = test_df['text_cleaned']
y_test_ml = test_df['label']

print("\nExtracting features using TF-IDF on cleaned text...")
tfidf_vectorizer = TfidfVectorizer(max_features=5000, ngram_range=(1, 2))
X_train_tfidf = tfidf_vectorizer.fit_transform(X_train_ml)
X_test_tfidf = tfidf_vectorizer.transform(X_test_ml)
print(f"✅ TF-IDF matrix shape: {X_train_tfidf.shape}")

# Definisikan model
ml_models = {
    "Naive Bayes": MultinomialNB(),
    "SVM (Linear)": SVC(kernel='linear'),
    "Random Forest": RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1)
}

# Latih dan evaluasi
for name, model in ml_models.items():
    print(f"\n--- Training {name} ---")
    
    # Training
    model.fit(X_train_tfidf, y_train_ml)
    
    # Inference
    start_infer = time.time()
    y_pred = model.predict(X_test_tfidf)
    inference_time = time.time() - start_infer
    
    # Evaluasi
    print(f"✅ {name} Evaluation Report:")
    report = classification_report(y_test_ml, y_pred, target_names=emotion_to_id.keys(), digits=4)
    print(report)
    print(f"Inference Time: {inference_time:.4f} seconds")
    
    # Confusion Matrix
    cm = confusion_matrix(y_test_ml, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=emotion_to_id.keys(), yticklabels=emotion_to_id.keys())
    plt.title(f'Confusion Matrix - {name}')
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.show()

print("-" * 50)

# ===== 3. Model Deep Learning - BiLSTM =====
print("\n===== 3. Training BiLSTM Model =====")

# Hyperparameters BiLSTM
VOCAB_SIZE = 10000
EMBEDDING_DIM = 128
MAX_LEN = 128
LEARNING_RATE_BILSTM = 1e-3
EPOCHS_BILSTM = 20
BATCH_SIZE_BILSTM = 32 # Batch size yang wajar untuk BiLSTM

# --- Persiapan Data untuk BiLSTM ---
# Tokenisasi menggunakan Keras Tokenizer untuk kontrol vocab_size
bilstm_tokenizer = Tokenizer(num_words=VOCAB_SIZE, oov_token="<unk>")
bilstm_tokenizer.fit_on_texts(train_df['text_cleaned'])

train_seq = bilstm_tokenizer.texts_to_sequences(train_df['text_cleaned'])
val_seq = bilstm_tokenizer.texts_to_sequences(val_df['text_cleaned'])
test_seq = bilstm_tokenizer.texts_to_sequences(test_df['text_cleaned'])

# Padding
X_train_pad = pad_sequences(train_seq, maxlen=MAX_LEN, padding='post', truncating='post')
X_val_pad = pad_sequences(val_seq, maxlen=MAX_LEN, padding='post', truncating='post')
X_test_pad = pad_sequences(test_seq, maxlen=MAX_LEN, padding='post', truncating='post')

# Konversi ke PyTorch DataLoader
class BiLSTMDataset(Dataset):
    def __init__(self, sequences, labels):
        self.sequences = torch.tensor(sequences, dtype=torch.long)
        self.labels = torch.tensor(labels, dtype=torch.long)
    def __len__(self): return len(self.sequences)
    def __getitem__(self, idx): return self.sequences[idx], self.labels[idx]

train_loader_bilstm = DataLoader(BiLSTMDataset(X_train_pad, train_df['label'].values), batch_size=BATCH_SIZE_BILSTM, shuffle=True)
val_loader_bilstm = DataLoader(BiLSTMDataset(X_val_pad, val_df['label'].values), batch_size=BATCH_SIZE_BILSTM)
test_loader_bilstm = DataLoader(BiLSTMDataset(X_test_pad, test_df['label'].values), batch_size=BATCH_SIZE_BILSTM)

# --- Model BiLSTM di PyTorch ---
class BiLSTMClassifier(nn.Module):
    def __init__(self, vocab_size, embedding_dim, hidden_dim, output_dim, n_layers=2, dropout=0.5):
        super().__init__()
        self.embedding = nn.Embedding(vocab_size, embedding_dim, padding_idx=0)
        self.lstm = nn.LSTM(embedding_dim, hidden_dim, num_layers=n_layers, 
                              bidirectional=True, dropout=dropout if n_layers > 1 else 0, batch_first=True)
        self.fc = nn.Linear(hidden_dim * 2, output_dim)
        self.dropout = nn.Dropout(dropout)
        
    def forward(self, text):
        embedded = self.dropout(self.embedding(text))
        outputs, (hidden, cell) = self.lstm(embedded)
        hidden = self.dropout(torch.cat((hidden[-2,:,:], hidden[-1,:,:]), dim=1))
        return self.fc(hidden)

model_bilstm = BiLSTMClassifier(VOCAB_SIZE, EMBEDDING_DIM, 128, num_labels).to(device)
optimizer_bilstm = torch.optim.Adam(model_bilstm.parameters(), lr=LEARNING_RATE_BILSTM)
criterion = nn.CrossEntropyLoss().to(device)

def train_eval_bilstm(model, train_loader, val_loader, optimizer, criterion, epochs, model_name):
    history = {'train_loss': [], 'train_acc': [], 'val_loss': [], 'val_acc': []}
    best_val_acc = 0
    
    for epoch in range(epochs):
        model.train()
        total_loss, total_correct = 0, 0
        for seq, labels in train_loader:
            seq, labels = seq.to(device), labels.to(device)
            optimizer.zero_grad()
            output = model(seq)
            loss = criterion(output, labels)
            loss.backward()
            optimizer.step()
            
            total_loss += loss.item()
            preds = output.argmax(dim=1)
            total_correct += (preds == labels).sum().item()
        
        train_loss = total_loss / len(train_loader)
        train_acc = total_correct / len(train_loader.dataset)
        history['train_loss'].append(train_loss)
        history['train_acc'].append(train_acc)

        model.eval()
        val_loss, val_correct = 0, 0
        with torch.no_grad():
            for seq, labels in val_loader:
                seq, labels = seq.to(device), labels.to(device)
                output = model(seq)
                loss = criterion(output, labels)
                val_loss += loss.item()
                preds = output.argmax(dim=1)
                val_correct += (preds == labels).sum().item()
        
        val_loss /= len(val_loader)
        val_acc = val_correct / len(val_loader.dataset)
        history['val_loss'].append(val_loss)
        history['val_acc'].append(val_acc)
        
        print(f"Epoch {epoch+1}/{epochs} | Train Loss: {train_loss:.4f} Acc: {train_acc:.4f} | Val Loss: {val_loss:.4f} Acc: {val_acc:.4f}")
        
        if val_acc > best_val_acc:
            best_val_acc = val_acc
            torch.save(model.state_dict(), f'{model_name}_best.pt')

    return history

print("\n--- Training BiLSTM ---")
history_bilstm = train_eval_bilstm(model_bilstm, train_loader_bilstm, val_loader_bilstm, optimizer_bilstm, criterion, EPOCHS_BILSTM, 'bilstm')

# --- Evaluasi BiLSTM ---
model_bilstm.load_state_dict(torch.load('bilstm_best.pt'))
model_bilstm.eval()
y_pred_list, y_true_list = [], []

start_infer = time.time()
with torch.no_grad():
    for seq, labels in test_loader_bilstm:
        seq = seq.to(device)
        output = model_bilstm(seq)
        y_pred_list.extend(output.argmax(dim=1).cpu().numpy())
        y_true_list.extend(labels.numpy())
inference_time = time.time() - start_infer

print("\n✅ BiLSTM Evaluation Report:")
print(classification_report(y_true_list, y_pred_list, target_names=emotion_to_id.keys(), digits=4))
print(f"Inference Time: {inference_time:.4f} seconds")

# Plotting
def plot_history(history, model_name):
    plt.figure(figsize=(14, 5))
    plt.subplot(1, 2, 1)
    plt.plot(history['train_loss'], label='Train Loss')
    plt.plot(history['val_loss'], label='Validation Loss')
    plt.title(f'{model_name} Loss')
    plt.legend()
    plt.subplot(1, 2, 2)
    plt.plot(history['train_acc'], label='Train Accuracy')
    plt.plot(history['val_acc'], label='Validation Accuracy')
    plt.title(f'{model_name} Accuracy')
    plt.legend()
    plt.show()

plot_history(history_bilstm, "BiLSTM")
cm = confusion_matrix(y_true_list, y_pred_list)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=emotion_to_id.keys(), yticklabels=emotion_to_id.keys())
plt.title('Confusion Matrix - BiLSTM')
plt.show()

print("-" * 50)


# ===== 4. Model Deep Learning - Transformers =====
print("\n===== 4. Training Transformer Models =====")

# Hyperparameters Transformers
MAX_LEN_TRANS = 128
BATCH_SIZE = 16
EPOCHS_TRANS = 10
LEARNING_RATE_TRANS = 2e-5

# Dataset dan DataLoader untuk Transformers
class TransformerDataset(Dataset):
    def __init__(self, texts, labels, tokenizer, max_len):
        self.texts = texts
        self.labels = labels
        self.tokenizer = tokenizer
        self.max_len = max_len
    
    def __len__(self): return len(self.texts)
    
    def __getitem__(self, item):
        text = str(self.texts[item])
        label = self.labels[item]
        encoding = self.tokenizer.encode_plus(
            text,
            add_special_tokens=True,
            max_length=self.max_len,
            return_token_type_ids=False,
            padding='max_length',
            truncation=True,
            return_attention_mask=True,
            return_tensors='pt',
        )
        return {
            'input_ids': encoding['input_ids'].flatten(),
            'attention_mask': encoding['attention_mask'].flatten(),
            'labels': torch.tensor(label, dtype=torch.long)
        }

def create_trans_loader(df, tokenizer, max_len, batch_size, shuffle=False):
    ds = TransformerDataset(
        texts=df.text_cleaned.to_numpy(),
        labels=df.label.to_numpy(),
        tokenizer=tokenizer,
        max_len=max_len
    )
    return DataLoader(ds, batch_size=batch_size, shuffle=shuffle, num_workers=2)

def train_eval_transformer(model, train_loader, val_loader, optimizer, scheduler, epochs, model_name):
    history = {'train_loss': [], 'train_acc': [], 'val_loss': [], 'val_acc': []}
    best_val_acc = 0

    for epoch in range(epochs):
        model.train()
        total_loss, total_correct = 0, 0
        for batch in train_loader:
            input_ids = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            labels = batch['labels'].to(device)
            
            optimizer.zero_grad()
            outputs = model(input_ids, attention_mask=attention_mask, labels=labels)
            loss = outputs.loss
            loss.backward()
            optimizer.step()
            scheduler.step()
            
            total_loss += loss.item()
            preds = outputs.logits.argmax(dim=1)
            total_correct += (preds == labels).sum().item()
            
        train_loss = total_loss / len(train_loader)
        train_acc = total_correct / len(train_loader.dataset)
        history['train_loss'].append(train_loss)
        history['train_acc'].append(train_acc)
        
        model.eval()
        val_loss, val_correct = 0, 0
        with torch.no_grad():
            for batch in val_loader:
                input_ids = batch['input_ids'].to(device)
                attention_mask = batch['attention_mask'].to(device)
                labels = batch['labels'].to(device)
                outputs = model(input_ids, attention_mask=attention_mask, labels=labels)
                val_loss += outputs.loss.item()
                val_correct += (outputs.logits.argmax(dim=1) == labels).sum().item()
        
        val_loss /= len(val_loader)
        val_acc = val_correct / len(val_loader.dataset)
        history['val_loss'].append(val_loss)
        history['val_acc'].append(val_acc)
        
        print(f"Epoch {epoch+1}/{epochs} | Train Loss: {train_loss:.4f} Acc: {train_acc:.4f} | Val Loss: {val_loss:.4f} Acc: {val_acc:.4f}")
        
        if val_acc > best_val_acc:
            best_val_acc = val_acc
            model.save_pretrained(f'./{model_name}_best')
            tokenizer.save_pretrained(f'./{model_name}_best')
    
    return history

# Looping untuk setiap model Transformer
transformer_models = {
    "BERT": (BertForSequenceClassification, BertTokenizer, 'bert-base-uncased'),
    "RoBERTa": (RobertaForSequenceClassification, RobertaTokenizer, 'roberta-base'),
    "DistilBERT": (DistilBertForSequenceClassification, DistilBertTokenizer, 'distilbert-base-uncased')
}

for name, (ModelClass, TokenizerClass, model_name) in transformer_models.items():
    print(f"\n--- Processing {name} ---")
    
    # Inisialisasi Tokenizer dan DataLoader
    tokenizer = TokenizerClass.from_pretrained(model_name)
    train_loader = create_trans_loader(train_df, tokenizer, MAX_LEN_TRANS, BATCH_SIZE, shuffle=True)
    val_loader = create_trans_loader(val_df, tokenizer, MAX_LEN_TRANS, BATCH_SIZE)
    test_loader = create_trans_loader(test_df, tokenizer, MAX_LEN_TRANS, BATCH_SIZE)
    
    # Inisialisasi Model, Optimizer, Scheduler
    model = ModelClass.from_pretrained(model_name, num_labels=num_labels).to(device)
    optimizer = AdamW(model.parameters(), lr=LEARNING_RATE_TRANS)
    total_steps = len(train_loader) * EPOCHS_TRANS
    scheduler = get_linear_schedule_with_warmup(optimizer, num_warmup_steps=0, num_training_steps=total_steps)
    
    # Training
    history = train_eval_transformer(model, train_loader, val_loader, optimizer, scheduler, EPOCHS_TRANS, name)
    
    # Evaluasi
    print(f"\n✅ {name} Evaluation Report:")
    best_model = ModelClass.from_pretrained(f'./{name}_best').to(device)
    best_model.eval()
    
    y_pred_list, y_true_list = [], []
    start_infer = time.time()
    with torch.no_grad():
        for batch in test_loader:
            input_ids = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            labels = batch['labels']
            outputs = best_model(input_ids, attention_mask=attention_mask)
            y_pred_list.extend(outputs.logits.argmax(dim=1).cpu().numpy())
            y_true_list.extend(labels.numpy())
    inference_time = time.time() - start_infer

    print(classification_report(y_true_list, y_pred_list, target_names=emotion_to_id.keys(), digits=4))
    print(f"Inference Time: {inference_time:.4f} seconds")
    
    save_prediction_results(test_df, y_true_list, y_pred_list, id_to_emotion, name)
    
    plot_history(history, name)
    cm = confusion_matrix(y_true_list, y_pred_list)
    plt.figure(figsize=(8, 6)); sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=emotion_to_id.keys(), yticklabels=emotion_to_id.keys()); plt.title(f'Confusion Matrix - {name}'); plt.show()


print("\n\n===== All models have been trained and evaluated. =====")
```
